## Transformada de fourier na verdade transforma o tempo em frequencia

```python
import numpy as  np
import matplotlib.pyplot as plt
from scipy.fftpack import fft #scipy é um pacote cientifico do numpy
%matplotlib inline
# Number of sample points
N = 600 #Amostras
# sample spacing
T = 1.0 / 800.0 #Esse é  o T
x = np.linspace(0.0, N*T, N)
y = np.sin(200.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)#  x mede  o tempo / possui duas frequencias, a priemira 2pí50x  50=f'  e  na segunda 2pí80x 80=f² 
plt.grid()
plt.plot(x,y)
plt.show()
yf = fft(y)#FFT é uma função que se chama, transformada de Fourier - fast fourier transform
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)# aqui mostra o vetor das frequencias (0 até 1/(2*t))=(1/2)*(1/T)=> 1/2*f(Frequecia)
import matplotlib.pyplot as plt
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```

## 1.pediu para fazer a fft em cima de um seno de t /600

```python
import matplotlib.pyplot as plt
%matplotlib inline
t = np.arange(1024)
sp = np.fft.fft(np.sin(t))/600 # Aqui 1. 
freq = np.fft.fftfreq(t.shape[-1])/600
plt.plot(freq, sp.real,freq,sp.imag) # Aqui faz a parte real e a imaginaria/parte real para cima e parte imaginaria para baixo
plt.show()
```
