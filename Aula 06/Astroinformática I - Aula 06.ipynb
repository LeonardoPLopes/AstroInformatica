```python
import matplotlib.pyplot as plt
import numpy as np
t=np.arange(0.,7.,0.2)# aqui mostra os 3 / o terceiro mostra o espaçamento(0.2)/primeira  e  segunda marcação mostra o X e o Y dografico
plt.plot(t,t,"r--",t,t**2,"bs",t,t**3,"g^")#Foi tudo feito no mesmo plot é de 3 em 3
plt.show()
```

## color- voce pode  colocar  a  cor da  linha, aceita até em hexadecimal
##  Linewidth - faz  a grossura da linha
## linestyle - é  para  colocar como vc quer q  apareça a linha(quadrada ou tracejada)
## label - ele faz um grafico  

### plt.legend(loc="upper left")  pode tambem colocar  rigth(direita) ou center ou left (esquerda)

```python
import matplotlib.pyplot as plt
import numpy as np

t=np.arange(0.,5.,0.2)
plt.plot(t,t,color="red",linewidth=4,label="Temp C")
plt.plot(t,t,"ro");
plt.plot(t,t**2,color="blue",linewidth=7,label="Temp F")
plt.plot(t,t**3,color="green",linewidth=1,label="Temp K")
plt.legend(loc="upper left")#Aqui ele coloca o grafico no local onde voce deseja.
plt.show()
```

## alpha = transparencia de 0,75
## normed=normalizado de 0 até 1

```python
import matplotlib.pyplot as plt
import numpy as np

mu, sigma = 100,15 # media 100 desvio padrão 15
x=mu+sigma*np.random.randn(10000)  # 10000 numeros aleatorios que é feito vezes o sigma mais o mí (lembrem de analise de dados kkkk)
n,bins,ptaches=plt.hist(x,50,normed=1,facecolor="g",alpha=0.75)# função para fazer o estograma 
plt.xlabel("Smarts")#Titulo abaixo do grafico
plt.ylabel("Probability")# titulo ao lado do grafico
plt.title("Histogram of IQ")# titulo acima do grafico
plt.text(60,0.025,r"$\mu=100,\ \sigma=15$")#  aqui ele faz a letra grega mí e tambem o sigma (o texto a esqueda do grafico)/ posição y=60 e X=0,025
plt.axis([40,160,0,0.03])# aqui é  o tamanho do grafico
plt.grid(True)# aqui faz a gradezinha do grafico
plt.show()
```
